Series.
<br>
</h2>

<div style="font-family: 'Roboto', sans-serif; font-weight: 400; text-align: justify; font-size: 1.1em; line-height: 1.6;">
Una Series es un arreglo unidimensional con índices.
    <br>
</div>

```python
"""
Este código muestra cómo crear y manipular objetos Series con la librería pandas, 
una estructura fundamental para trabajar con datos unidimensionales. 
A través de ejemplos prácticos, se exploran la creación, indexación personalizada y operaciones básicas, 
lo que es clave para realizar análisis de datos eficientes y claros.
"""

# Importa la biblioteca pandas y la asigna al alias 'pd' para facilitar su uso
import pandas as pd

# Crear una Series a partir de una lista simple de números
s = pd.Series([10, 20, 30, 40])

# Mostrar la Series completa en consola para visualizar su contenido
print(s)

# Acceder al primer elemento de la Series usando índice numérico
print(s[0])  # Imprime el valor en la posición 0 (primer elemento)

# Realizar un slicing para obtener un subconjunto de la Series (elementos en las posiciones 1 y 2)
print(s[1:3])  # Aquí el rango es inclusivo en el inicio y exclusivo en el final

# Crear una Series con índices personalizados (etiquetas en lugar de posiciones numéricas)
s2 = pd.Series([100, 200, 300], index=['a', 'b', 'c'])

# Acceder a un elemento usando su índice personalizado 'b'
print(s2['b'])  # Devuelve el valor asociado a la etiqueta 'b'

# Sumar 10 a cada elemento de la Series 's', demostrando operaciones vectorizadas
print(s + 10)  # Esto mejora la legibilidad y eficiencia frente a un loop manual

# Calcular el promedio (media aritmética) de todos los valores en la Series 's'
print(s.mean())  # Muy útil para obtener resúmenes estadísticos rápidos
  

"""
En conclusión, este código ejemplifica cómo crear Series, acceder a sus datos con índices numéricos y personalizados,
y realizar operaciones matemáticas básicas con pandas. Estas habilidades son esenciales para manipular y analizar datos
de forma rápida, intuitiva y con un código más limpio y eficiente.
"""
```

Un DataFrame es una tabla bidimensional con filas y columnas etiquetadas.

```python
"""
Este código demuestra cómo crear DataFrames con la biblioteca pandas a partir de diferentes estructuras de datos, 
como un diccionario y una lista de listas. Los DataFrames son estructuras tabulares muy poderosas para el análisis 
y manipulación de datos, fundamentales en muchos proyectos de ciencia de datos y análisis en Python.
"""

# Importar la biblioteca pandas con el alias 'pd' para trabajar con DataFrames
import pandas as pd

# Definir un diccionario con claves que serán los nombres de las columnas y valores como listas que representan filas
data = {
    'Nombre': ['Ana', 'Luis', 'Pedro'],  # Columna "Nombre" con tres nombres de personas
    'Edad': [23, 34, 45],                # Columna "Edad" con edades correspondientes
    'Ciudad': ['Madrid', 'Valencia', 'Sevilla']  # Columna "Ciudad" con ciudades respectivas
}

# Crear un DataFrame a partir del diccionario 'data'
# Pandas interpreta las claves como nombres de columnas y las listas como valores fila a fila
df = pd.DataFrame(data)

# Mostrar el DataFrame creado para verificar su contenido y estructura
print(df)

# Definir una lista de listas, donde cada lista interior representa una fila de dos elementos
data2 = [[1, 2], [3, 4], [5, 6]]

# Crear un DataFrame usando la lista 'data2' y especificando los nombres de las columnas
# Esto es útil si se tienen datos tabulares simples sin etiquetas previas
df2 = pd.DataFrame(data2, columns=['A', 'B'])

# Mostrar el segundo DataFrame para comprobar cómo se asignaron las columnas y filas
print(df2)

"""
En resumen, este código muestra cómo generar DataFrames en pandas de forma sencilla a partir de distintos tipos de datos. 
Esta flexibilidad es crucial para alimentar análisis y procesamiento de datos estructurados desde múltiples fuentes.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Valores faltantes (NaN).
<br>
</h2>

```python
"""
Este código ejemplifica cómo manejar valores nulos (NaN) en un DataFrame de pandas rellenándolos con la media de cada columna. 
Tratar los valores faltantes es una etapa crucial en la preparación de datos, ya que permite mantener la integridad y continuar con análisis o modelados sin errores. 
Rellenar con la media es una técnica común para imputar datos que ayuda a conservar la distribución original.
"""

# Importar la biblioteca pandas para manejo de datos y numpy para funciones matemáticas avanzadas
import pandas as pd
import numpy as np

# Crear un DataFrame con dos columnas 'A' y 'B', que incluye valores numéricos y algunos NaN (valores faltantes)
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],      # Columna A con un NaN en la tercera posición
    'B': [5, np.nan, np.nan, 8, 10] # Columna B con dos NaN en la segunda y tercera posición
})

# Mostrar el DataFrame original que contiene valores nulos para observar la situación inicial
print("DataFrame original con valores nulos:")
print(df)

# Rellenar los valores NaN de cada columna con la media (promedio) correspondiente a esa columna
# df.mean() calcula la media ignorando los NaN automáticamente
# fillna() reemplaza los NaN con esos valores calculados
df_filled = df.fillna(df.mean())

# Mostrar el DataFrame resultante después de reemplazar los valores faltantes con las medias
print("\nDataFrame después de rellenar NaN con la media de cada columna:")
print(df_filled)

"""
En conclusión, este código ilustra una forma sencilla y efectiva de imputar valores faltantes en un DataFrame usando la media. 
Esto ayuda a completar datos para análisis más fiables y evita problemas derivados de datos incompletos.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Manejo de datos atípicos usando la media.
<br>
</h2>

<div style="font-family: 'Roboto', sans-serif; font-weight: 400; text-align: justify; font-size: 1.1em; line-height: 1.6;">
Los datos atípicos (outliers) son valores que difieren significativamente de la mayoría de los datos. Una forma simple de manejarlos es reemplazarlos por la media si se detectan fuera de un rango razonable, por ejemplo usando el método del rango intercuartílico (IQR).
    <br>
</div>

```python
"""
Este código define una función para detectar y reemplazar valores atípicos (outliers) en una columna específica de un DataFrame con la media de esa columna. 
Eliminar o corregir outliers es crucial para evitar que estos datos extremos distorsionen análisis estadísticos y modelos predictivos. 
El método usado aquí se basa en el rango intercuartílico (IQR), un enfoque común para identificar valores fuera de lo esperado.
"""

# Importar numpy para realizar operaciones numéricas, principalmente para la función np.where
import numpy as np
import pandas as pd

# Función que recibe un DataFrame y el nombre de columna donde se detectarán outliers para reemplazarlos con la media
def reemplazar_outliers_por_media(df, columna):
    # Calcular el primer cuartil (Q1), que es el valor debajo del cual se encuentra el 25% de los datos
    Q1 = df[columna].quantile(0.25)
    
    # Calcular el tercer cuartil (Q3), que es el valor debajo del cual está el 75% de los datos
    Q3 = df[columna].quantile(0.75)
    
    # Calcular el rango intercuartílico (IQR), diferencia entre Q3 y Q1
    IQR = Q3 - Q1
    
    # Definir el límite inferior para identificar outliers (valores muy bajos)
    limite_inferior = Q1 - 1.5 * IQR
    
    # Definir el límite superior para identificar outliers (valores muy altos)
    limite_superior = Q3 + 1.5 * IQR
    
    # Calcular la media de la columna para usarla como reemplazo cuando se detecte un outlier
    media_col = df[columna].mean()
    
    # Usar np.where para crear una nueva columna donde se reemplazan valores fuera de los límites por la media
    # Se mantiene el valor original si está dentro de los límites
    df[columna] = np.where(
        (df[columna] < limite_inferior) | (df[columna] > limite_superior),
        media_col,
        df[columna]
    )
    
    # Devolver el DataFrame modificado con los outliers reemplazados
    return df

# Crear un DataFrame de ejemplo con una columna 'Edad' y un valor claramente atípico (120)
df_outliers = pd.DataFrame({
    'Edad': [23, 22, 21, 120, 24, 26, 27]  # 120 es un outlier evidente
})

# Mostrar el DataFrame original que contiene un valor outlier para observar el punto problemático
print("\nDataFrame original con outlier:")
print(df_outliers)

# Aplicar la función para reemplazar outliers en la columna 'Edad'
df_limpio = reemplazar_outliers_por_media(df_outliers, 'Edad')

# Mostrar el DataFrame resultante donde el valor outlier ha sido sustituido por la media de la columna
print("\nDataFrame después de reemplazar outliers con la media:")
print(df_limpio)

"""
En resumen, este código identifica y corrige valores extremos en los datos para mejorar la calidad del DataFrame. 
Reemplazar outliers con la media ayuda a reducir el impacto de valores atípicos en análisis estadísticos y modelos posteriores, 
manteniendo la estructura completa del conjunto de datos.
"""
```

```python
"""
Este código demuestra cómo cambiar los nombres de las columnas y los índices de un DataFrame en pandas. 
Renombrar columnas e índices facilita la comprensión y claridad de los datos, 
lo cual es especialmente útil para la presentación y manipulación posterior del DataFrame.
"""

# Cambiar el nombre de la columna 'Nombre' a 'NombreCompleto' directamente en el DataFrame original
# Se usa inplace=True para modificar el DataFrame sin necesidad de reasignarlo a una nueva variable
df.rename(columns={'Nombre': 'NombreCompleto'}, inplace=True)

# Cambiar la etiqueta del índice 0 a 'primera_fila' en el DataFrame original
# También se usa inplace=True para aplicar el cambio directamente en el DataFrame existente
df.rename(index={0: 'primera_fila'}, inplace=True)

"""
En resumen, estos comandos permiten actualizar etiquetas de columnas e índices para hacer el DataFrame más intuitivo y legible. 
Esto ayuda a evitar confusiones y facilita el trabajo con los datos en análisis posteriores.
"""
```

Filtrar, ordenar y clasificar datos.

</h2>

```python
"""
Este código realiza operaciones comunes de manipulación de datos en un DataFrame de pandas: 
filtra filas según una condición específica y ordena los datos de forma descendente por una columna. 
Estas técnicas son fundamentales para extraer información relevante y organizar los datos para su análisis más efectivo.
"""

# Filtrar el DataFrame 'df' para obtener solo las filas donde el valor de la columna 'Edad' sea mayor a 25
# Esto genera un nuevo DataFrame 'df_filtered' con los datos que cumplen esa condición
df_filtered = df[df['Edad'] > 25]

# Ordenar el DataFrame original 'df' según los valores de la columna 'Edad' en orden descendente (de mayor a menor)
# El resultado se guarda en un nuevo DataFrame 'df_sorted'
df_sorted = df.sort_values(by='Edad', ascending=False)

# Mostrar el DataFrame ordenado 'df_sorted' para visualizar los resultados
print(df_sorted)

"""
En conclusión, este fragmento permite extraer subconjuntos específicos de datos y ordenarlos para facilitar el análisis y la interpretación. 
Estas operaciones son clave en el procesamiento de datos para obtener insights relevantes y presentarlos claramente.
"""
```

Duplicados.
<br>
</h2>

```python
"""
Este código detecta y elimina filas duplicadas en un DataFrame de pandas. 
Eliminar duplicados es fundamental para limpiar datos y asegurar que el análisis posterior sea preciso y fiable, 
evitando sesgos o errores causados por información repetida.
"""

# Detectar filas duplicadas en el DataFrame 'df'
# La función duplicated() devuelve una Serie booleana que indica True para las filas que son duplicadas respecto a una anterior
print(df.duplicated())

# Eliminar todas las filas duplicadas del DataFrame 'df'
# drop_duplicates() devuelve un nuevo DataFrame sin las filas repetidas, manteniendo únicamente la primera aparición de cada duplicado
df_no_duplicates = df.drop_duplicates()

"""
En resumen, este código facilita la identificación y eliminación de datos repetidos en un DataFrame. 
Esta limpieza es clave para mantener la calidad de los datos y mejorar la exactitud de los resultados en cualquier análisis posterior.
"""
```

Conversión de tipos.
<br>
</h2>

```python
"""
Este código convierte una columna específica de un DataFrame de pandas al tipo de dato entero (int). 
Realizar conversiones de tipo es fundamental para asegurar que los datos tienen el formato correcto antes de análisis o procesamiento, 
lo que evita errores y garantiza la coherencia en los cálculos y operaciones posteriores.
"""

# Convertir la columna 'Edad' del DataFrame 'df' al tipo entero
# astype(int) fuerza la conversión de los valores, útil cuando se quiere trabajar con números enteros en lugar de decimales o strings
df['Edad'] = df['Edad'].astype(int)

"""
En resumen, esta línea asegura que los datos de la columna 'Edad' estén en formato entero, 
lo que es esencial para realizar operaciones precisas y evitar problemas relacionados con tipos de datos inconsistentes.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Aplicar funciones.
<br>
</h2>

```python
"""
Este código muestra cómo aplicar funciones personalizadas a datos en un DataFrame de pandas mediante los métodos apply, map y applymap. 
Estas técnicas permiten transformar y manipular los datos de manera flexible y eficiente, adaptándose a diferentes necesidades según la estructura a modificar.
"""

# Crear una nueva columna 'EdadPlus10' sumando 10 a cada valor de la columna 'Edad'
# apply permite aplicar una función a cada elemento de la Serie (columna), en este caso una función lambda que suma 10
df['EdadPlus10'] = df['Edad'].apply(lambda x: x + 10)

# Reemplazar los valores de la columna 'Ciudad' usando un diccionario de mapeo
# map es ideal para Series y facilita la traducción o cambio selectivo de valores, por ejemplo aquí se abrevia el nombre de las ciudades
df['Ciudad'] = df['Ciudad'].map({'Madrid': 'M', 'Valencia': 'V'})

# Aplicar una función a cada elemento (celda) del DataFrame que contiene solo la columna 'Edad'
# applymap funciona a nivel elemento en todo el DataFrame, por eso se usa sobre un DataFrame (aquí una selección con solo 'Edad')
# La función lambda multiplica cada valor por 2
df_num = df[['Edad']].applymap(lambda x: x * 2)

"""
En resumen, estos métodos brindan flexibilidad para transformar datos en pandas: 
apply para aplicar funciones a Series, map para reemplazos específicos y applymap para operaciones elementales en todo un DataFrame. 
Saber elegir el método correcto optimiza el código y facilita la manipulación de datos según necesidades concretas.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Agrupar datos con groupby.
<br>
</h2>

```python
"""
Este código ilustra cómo agrupar datos en un DataFrame de pandas según una columna (en este caso, 'Ciudad') 
y calcular estadísticas resumidas de otra columna ('Edad'). 
Agrupar y agregar datos facilita el análisis por categorías, permitiendo obtener información importante para la toma de decisiones y reportes.
"""

# Agrupar el DataFrame por valores únicos en la columna 'Ciudad' y calcular el promedio de la columna 'Edad' para cada grupo
# Esto genera una Serie donde el índice son las ciudades y los valores, las medias correspondientes
grupo = df.groupby('Ciudad')['Edad'].mean()

# Mostrar los promedios de edad calculados para cada ciudad
print(grupo)

# Realizar agrupación por 'Ciudad' y aplicar múltiples funciones agregadas a la columna 'Edad'
# mean: promedio, max: valor máximo, min: valor mínimo de la edad por cada ciudad
agg = df.groupby('Ciudad').agg({'Edad': ['mean', 'max', 'min']})

# Mostrar el DataFrame resultante con las agregaciones múltiples para cada ciudad
print(agg)

"""
En conclusión, este código permite resumir datos mediante agrupamientos y estadísticas clave, 
facilitando la comprensión de patrones y diferencias entre categorías. 
Estas técnicas son muy útiles para análisis exploratorios y reportes claros.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Pivot tables y crosstabs.
<br>
</h2>

```python
"""
Este código muestra cómo resumir y explorar datos categóricos usando pandas con tablas dinámicas (pivot tables) y tablas cruzadas (crosstab). 
Estas herramientas permiten organizar datos de forma estructurada, facilitando la interpretación y el análisis de relaciones entre variables.
"""

# Crear una tabla dinámica que agrupa el DataFrame 'df' por la columna 'Ciudad'
# Calcula el promedio ('mean') de la columna 'Edad' para cada ciudad
pivot = df.pivot_table(index='Ciudad', values='Edad', aggfunc='mean')

# Mostrar la tabla dinámica resultante con las medias de edad por ciudad
print(pivot)

# Crear una tabla cruzada que cuenta la frecuencia de combinaciones entre 'Ciudad' y 'Nombre'
# Esto ayuda a identificar cuántas veces aparece cada nombre en cada ciudad
tabla = pd.crosstab(df['Ciudad'], df['Nombre'])

# Mostrar la tabla cruzada que resume la ocurrencia de nombres por ciudad
print(tabla)

"""
En resumen, las pivot tables y las tablas cruzadas facilitan resumir y visualizar relaciones entre variables categóricas y numéricas. 
Estas técnicas son valiosas para análisis exploratorios y para presentar datos de manera clara y concisa.
"""
```

<h2 style="font-family: 'Lato', sans-serif; font-weight: 600; text-align: justify; color: #3c5a80; font-weight: bold; text-shadow: 3px 3px 14px rgba(60, 70, 100, 0.34); font-weight: bold;">
Merge, join y concatenación​.
<br>
</h2>

```python
"""
Este código muestra cómo combinar dos DataFrames en pandas usando dos técnicas comunes: concatenación y merge. 
Estas operaciones son esenciales para integrar datos provenientes de diferentes fuentes o estructuras, facilitando un análisis más completo y coherente.
"""

# Concatenar los DataFrames 'df' y 'df2' verticalmente (uno debajo del otro)
# axis=0 indica que la concatenación es por filas, agregando las filas de df2 al final de df
df_concat = pd.concat([df, df2], axis=0)

# Realizar un merge (similar a un JOIN en bases de datos) entre los DataFrames 'df' y 'df2'
# Se unen filas donde el valor en la columna 'Edad' de df coincide con el valor en la columna 'A' de df2
# how='inner' indica que sólo se conservarán las filas con coincidencias en ambas tablas
df_merge = df.merge(df2, left_on='Edad', right_on='A', how='inner')

"""
En conclusión, concatenar y hacer merge son dos métodos poderosos para ampliar y relacionar conjuntos de datos en pandas.
Estas herramientas permiten construir datasets más ricos y prepararlos para análisis avanzados o integración con otras fuentes.
"""
```

Operaciones con fechas y tiempos.
<br>
</h2>

```python
"""
Este código muestra cómo convertir una columna de un DataFrame a tipo datetime y extraer componentes individuales como año, mes y día. 
Trabajar con fechas correctamente es fundamental para análisis temporales y para manipular datos basados en tiempo de manera eficiente y precisa.
"""

# Convertir la columna 'Fecha' a tipo datetime de pandas
# errors='coerce' convierte valores inválidos o mal formateados en NaT (Not a Time) en lugar de lanzar un error
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Extraer el año de cada fecha y guardarlo en una nueva columna 'Año'
df['Año'] = df['Fecha'].dt.year

# Extraer el mes de cada fecha y guardarlo en una nueva columna 'Mes'
df['Mes'] = df['Fecha'].dt.month

# Extraer el día de cada fecha y guardarlo en una nueva columna 'Día'
df['Día'] = df['Fecha'].dt.day

"""
En resumen, este código facilita la transformación de datos temporales y la extracción de componentes específicos que son útiles para análisis detallados por períodos.
Manejar fechas como objetos datetime permite aprovechar las funciones avanzadas que pandas ofrece para series temporales.
"""
```

Análisis de dataset de ventas.
<br>
</h2>

```python
"""
Este código procesa datos de ventas cargados desde un archivo CSV para generar un reporte que muestra los productos más vendidos según el total facturado. 
Realizar este análisis es esencial para entender el desempeño de productos y apoyar la toma de decisiones comerciales basadas en cifras reales.
"""

# Cargar los datos de ventas desde el archivo 'ventas.csv' a un DataFrame llamado 'ventas'
ventas = pd.read_csv('ventas.csv')

# Eliminar filas que tengan valores nulos en las columnas 'Cantidad' o 'Precio'
# Esto asegura que los cálculos posteriores no contengan errores por datos incompletos
ventas = ventas.dropna(subset=['Cantidad', 'Precio'])

# Calcular el total de venta para cada fila multiplicando la cantidad por el precio unitario
# Se crea una nueva columna 'Total' que representa el ingreso generado por esa venta
ventas['Total'] = ventas['Cantidad'] * ventas['Precio']

# Agrupar las ventas por el nombre del producto y sumar los totales para obtener el total vendido por cada producto
# reset_index() se usa para transformar el resultado agrupado en un DataFrame plano
ventas_por_producto = ventas.groupby('Producto')['Total'].sum().reset_index()

# Ordenar los productos según el monto total vendido de forma descendente (de mayor a menor)
# Esto facilita identificar los productos más rentables o populares
ventas_por_producto = ventas_por_producto.sort_values(by='Total', ascending=False)

# Mostrar en pantalla las primeras filas del reporte con los productos principales y sus totales de venta
print(ventas_por_producto.head())

# Guardar el reporte final en un archivo CSV llamado 'reporte_ventas.csv'
# index=False evita que se guarde la columna de índices generada automáticamente
ventas_por_producto.to_csv('reporte_ventas.csv', index=False)

"""
En conclusión, este código limpia y procesa datos de ventas para producir un resumen útil que permite identificar fácilmente los productos con mayores ingresos. 
Este análisis es fundamental para monitorizar el rendimiento y gestionar estratégicamente el inventario o la oferta comercial.
"""
```
